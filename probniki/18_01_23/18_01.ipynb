```python
# 19

def game(h1, h2, cur, goal):
    if h1 + h2 >= 231:
        return cur % 2 == goal % 2
    if cur == goal:
        return False

    steps = [game(h1 + 2, h2, cur + 1, goal),
             game(h1 * 2, h2, cur + 1, goal),
             game(h1, h2 + 2, cur + 1, goal),
             game(h1, h2 * 2, cur + 1, goal)]

    return any(steps) if (cur + 1) % 2 == goal % 2 else all(steps)


for h2 in range(1, 214):
    for goal in range(1, 5):
        if game(17, h2, 0, goal):
            print(h2, goal)
            break
```

```python
from functools import lru_cache

win = 231
h1 = 17


def moves(h):
    a, b = h
    return (a + 2, b), (a, b + 2), (a * 2, b), (a, b * 2)


@lru_cache(None)
def f(h):
    if sum(h) >= win:
        return "W"
    if any(f(x) == "W" for x in moves(h)):
        return "P1"
    if all(f(x) == "P1" for x in moves(h)):
        return "B1"
    if any(f(x) == "B1" for x in moves(h)):
        return "P2"
    if all(f(x) == "P1" or f(x) == "P2" for x in moves(h)):
        return "B2"


for s in range(1, 214):
    if f((h1, s)) is not None:
        print(s, f((h1, s)))

# 211, 53 105, 96
```

```python
# 17
with open('17.txt', 'r') as f:
    data = [int(n) for n in f.readlines()]

su = -10 ** 6
cnt = 0
for i in range(len(data) - 1):
    x, y = data[i], data[i + 1]

    if x % 2 == 0 and y % 2 != 0:
        if x % 4 == 0 and y % 11 == 0:
            cnt += 1
            su = max(su, x + y)

    if x % 2 != 0 and y % 2 == 0:
        if x % 11 == 0 and y % 4 == 0:
            cnt += 1
            su = max(su, x + y)

print(cnt, su)
```

```python
# 16
from functools import lru_cache
import sys

sys.setrecursionlimit(200000)


@lru_cache(None)
def f(n):
    if n >= 10000:
        return n
    if n % 2 == 0:
        return f(n + 1) + n ** 2 - 3 * (n - 1)
    if n % 2 != 0:
        return f(n + 2) + 5 * n - (n - 1)


print(f(90) - f(99))
```

```python
# 14
def dec_to_base(x, y):
    w = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    s = ''

    while x > 0:
        if x % y > 9:
            s += w[x % y - 10]
        else:
            s += str(x % y)
        x //= y
    return s[::-1]


x = 5 * 36 ** 7 + 6 ** 10 - 36
res = dec_to_base(x, 6)
print(res.count('5'))
```

```python
# 5
cnt = 0
for n in range(100, 201):
    mx = max(c for c in str(n))
    mn = min(c for c in str(n))
    sm = sum(map(int, str(n)))
    n3 = sm - int(mx) - int(mn)
    if mn != '0':
        smx = mx + str(n3)
        smn = mn + str(n3)
        res = int(smx) - int(smn)
        if res == 30:
            cnt += 1
    if mn == '0':
        smx = mx + str(n3)
        smn = str(n3) + mn
        res = int(smx) - int(smn)
        if res == 30:
            cnt += 1

print(cnt)
```
